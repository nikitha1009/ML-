```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
def gradient_descent(x, y):
  m_curr = b_curr = 0
  iterations = 110
  n = len(x)
  learning_rate = 0.01
  cost_history = []
  m_history = []
  b_history = []

  for i in range(iterations):
    y_predicted = m_curr * x + b_curr
    cost = (1/n) * sum([val**2 for val in (y - y_predicted)])
    md = -(2/n) * sum(x * (y - y_predicted))
    bd = -(2/n) * sum(y - y_predicted)
    m_curr = m_curr - learning_rate * md
    b_curr = b_curr - learning_rate * bd

    cost_history.append(cost)
    m_history.append(m_curr)
    b_history.append(b_curr)

  return m_curr, b_curr, cost_history, m_history, b_history
```

```python
x = np.array([1, 2, 3, 4, 5])
y = np.array([5, 7, 9, 11, 13])

m, b, cost_history, m_history, b_history = gradient_descent(x, y)
```

```python

m_values = np.linspace(0, 4, 100)
b_values = np.linspace(0, 4, 100)
M, B = np.meshgrid(m_values, b_values)


cost_values = np.zeros_like(M)
for i in range(M.shape[0]):
  for j in range(M.shape[1]):
    y_predicted = M[i, j] * x + B[i, j]
    cost_values[i, j] = (1/len(x)) * sum([val**2 for val in (y - y_predicted)])


fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(M, B, cost_values, cmap='viridis')
ax.set_xlabel('m')
ax.set_ylabel('b')
ax.set_zlabel('Cost')
ax.set_title('Gradient Descent Visualization')


ax.plot(m_history, b_history, cost_history, color='red', marker='o', markersize=3)

plt.show()
```
